```python
#Handling missing data or values
#scenario-1:
#what we do value miss in the numerical column.
#we have to calculate mean or average.we can place the avg value in the empty place.
#missing in numerical column we have to calculate mean or average of that column.
#replace the avg or mean value in the empty place.
#scenario-2:
#missing values in unnumerical column.we have to replace in empty value with
#high repeated value.
#scenario-3:
#having missing values in usage data set.removing empty value rows from the data.

import numpy as np #np is an intance of numpy
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
#sklearn is a big library. simpleimputer assign a new values in the place of misiing values.
#simpleimputer is a calss from sklearn.which is used to new values placed in the missing values.
#simpleimputer and sklear are packages.
#simpleimputer is a class which was imported from sklearn package.
#fit can do prepare the imputer what the action do by imputer.
#fit can pass instruction to the imputer.
```

```python
X
```

```python
#step-4:handling the categorical data
#when the number are collected in groups or categories.
#converting the categorical data in to numerical data.(numbers)
#onehotencoder is class which was imported from sklearn package
#onehotencoder is used to covert text to binary code (0,1)
#columntransformer is a class which was imported from sklearn package
#columntransformer is used to apply changes or transformer (updating the valuses present in the data is transformer)
#column transformer is a class which was used to apply the changes of numeric data in the place of categorical data
#Alexa-0,0,1
#encoder is a method which converts text to numeric

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
X=ct.fit_transform(X)

```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)

#encoder the output column.(OneHotEncoder is used input columns. used for output encoder)
#OneHotEncoder is apply for input column
#labelencoder is only for output column
```

```python
Y
```

```python
#step-5:slipt the data into training and testing data.
#train_test_slipt is class which was imported from sklearn.model_selection.
#It was used for slipting the data into training and testing.
#the output based on test size.

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#step-6:feature scaling
#scaling is apply standard division to the values
#analyse the pattren

from sklearn.preprocessing import StandardScaler
sc=StandardScaler() #sc is object of standardscaler class.
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
```

```python
#Logistic Regression Algorithm
#when the model performce classifications task we have apply logistic regression algorithm
#Linear Regression Algorithm
#when the model performce prediction task we have to apply linear regression algurithm
#step-7:apply algorithm to the data set

from sklearn.linear_model import LogisticRegression

classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```
